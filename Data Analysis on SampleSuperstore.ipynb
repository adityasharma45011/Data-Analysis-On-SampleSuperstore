```python
plt.figure(figsize=(8, 6))
sns.barplot(x= store_df['Sub-Category'].head(10), y= store_df['Sales'].head(10), color='yellow')
plt.xlabel('Category')
plt.xticks(rotation = 45)
plt.ylabel('Sales')
plt.title(" Sales and Category")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x= store_df['Sub-Category'].head(10), y= store_df['Profit'].head(10), color='red',marker='o')
plt.xlabel('Category')
plt.xticks(rotation = 45)
plt.ylabel('Profit')
plt.title(" Profit and Category")
plt.show()
```

```python
store_df = store_df.drop(['Postal Code'],axis=1)
```

```python
sns.pairplot(store_df, hue='Ship Mode')
```

```python
sns.pairplot(store_df, hue='Segment')
```

```python
sns.pairplot(store_df, hue='Category')
```

```python
store_df['State'].value_counts()
```

```python
store_df.hist(figsize=(10,10), bins=50)
plt.show()
```

```python
sns.lineplot(x='Discount', y='Profit', data=store_df)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity', y='Profit',label='Profit', data=store_df)
plt.show()
```

```python
store_df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color=['pink','purple'],
                                                               figsize=(8, 6))
plt.xticks(rotation=45)
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
store_df.groupby('Region')[['Profit', 'Sales']].sum().plot.bar(color=['purple','skyblue'], figsize=(8,6))
plt.ylabel("Profit/Loss and Sales")
```

```python
new_df = store_df.groupby('State')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
new_df.plot.bar(color=['red','blue'], figsize=(15, 8))
plt.title('Profit/Loss and Sales across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
prof_loss = store_df["Profit"]
plt.plot(prof_loss)
plt.title("Profit and Loss")
```

```python
sns.boxenplot(store_df['Category'],prof_loss)
```

```python
sns.boxenplot(store_df['Ship Mode'],prof_loss)
```

```python
new_df1 = store_df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
new_df1.plot.bar(color=['red','blue'], figsize=(15, 8))
plt.title('Profit/Loss and Sales across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
